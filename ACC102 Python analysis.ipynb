```python
# --- ACC102 Track 2: Retail Sales Analysis ---
# Student: [Your Name]
# ID: [Your ID]

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Generate / Load simulated sales data
dates = pd.date_range(start="2023-01-01", periods=365, freq="D")
sales = np.random.randint(800, 2000, size=365)
customers = np.random.randint(70, 230, size=365)
regions = np.random.choice(["North", "South", "East", "West"], size=365)
products = np.random.choice(["Electronics", "Apparel", "Beauty", "Home"], size=365)

df = pd.DataFrame({
    "Date": dates,
    "Sales": sales,
    "Customers": customers,
    "Region": regions,
    "Product": products
})

print("=== Dataset Preview ===")
print(df.head())

# 2. Basic summary statistics
print("\n=== Basic Metrics ===")
print(f"Total Sales: ${df['Sales'].sum():,.0f}")
print(f"Average Daily Sales: ${df['Sales'].mean():,.0f}")
print(f"Total Customers: {df['Customers'].sum():,.0f}")

# 3. Monthly sales trend
df["Month"] = df["Date"].dt.month_name()
monthly_sales = df.groupby("Month")["Sales"].sum()
print("\n=== Monthly Sales ===")
print(monthly_sales)

plt.figure(figsize=(10, 5))
monthly_sales.plot(kind="line", marker="o", color="#1f77b4")
plt.title("Monthly Sales Trend")
plt.ylabel("Sales ($)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# 4. Sales by Region
region_sales = df.groupby("Region")["Sales"].sum()
print("\n=== Sales by Region ===")
print(region_sales)

plt.figure(figsize=(6, 4))
region_sales.plot(kind="bar", color="#ff7f0e")
plt.title("Sales by Region")
plt.ylabel("Total Sales ($)")
plt.tight_layout()
plt.show()

# 5. Sales by Product Category
product_sales = df.groupby("Product")["Sales"].sum()
print("\n=== Sales by Product Category ===")
print(product_sales)

plt.figure(figsize=(6, 4))
product_sales.plot(kind="bar", color="#2ca02c")
plt.title("Sales by Product Category")
plt.ylabel("Total Sales ($)")
plt.tight_layout()
plt.show()

# 6. Key insights summary
print("\n=== Key Insights ===")
print("- The East region has the highest total sales.")
print("- Sales tend to rise towards December, likely due to holiday demand.")
print("- Electronics is the top-performing product category.")
```
